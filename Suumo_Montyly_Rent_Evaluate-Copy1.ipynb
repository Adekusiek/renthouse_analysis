```python
import pandas as pd
from sklearn.externals import joblib

df = pd.read_csv("suumo_treated.csv", sep='\t', encoding='utf-16', index_col=0)
clf = joblib.load('suumo_rent_monthly_rent_station.pkl')

df = df.drop([29926, 69267, 69266, 199848, 76041])
df = df.drop(df.index[df['築年数'] == 99])
df = df.dropna()

df = df[df['駅'].str.contains('.*駅')]
station = pd.get_dummies(df['駅'])

df = pd.merge(df, station, left_index=True, right_index=True)

Y_true = df['賃料月額']
df.drop(['マンション名','区','路線', '駅', '賃料','管理費','敷金','礼金','保証金','敷引,償却', '賃料月額', '初期費用', '年間費用'], axis=1, inplace=True)  
data = df
```

```python
delta = clf.predict(data) - Y_true
delta.name = '予想賃料月額'
delta.head()
```

```python
df = pd.read_csv("suumo_treated.csv", sep='\t', encoding='utf-16', index_col=0)
df = df.drop([29926, 69267, 69266, 199848, 76041])
df = df.drop(df.index[df['築年数'] == 99])
df = df.dropna()

df = pd.concat([df, delta], axis=1)
#df = df[df['駅']=='牛込神楽坂駅']
df = df.sort_values(by='予想賃料月額', ascending=False)
df.head(40)
```
