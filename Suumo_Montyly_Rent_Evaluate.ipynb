```python
import pandas as pd
from sklearn.externals import joblib

df_origin = pd.read_csv("suumo_treated_DB.csv",sep = '\t', index_col=0)
df = df_origin
clf = joblib.load('suumo_rent_monthly_rent.pkl')

df = df.drop([29706, 5414])
df = df.drop(df.index[df['age'] == 99])
df = df.drop(df.index[df['floor'] < -2 ])
df = df.drop(df.index[df['floor_plan'] > 9 ])
df = df.drop(df.index[(df['floor_plan'] < 3) & (df['surface'] > 400 )])
df = df.drop(df.index[(df['walk_min1'] > 40) ])

ward = pd.get_dummies(df['ward'])

df_o = df[['floor_plan', 'plan_DK','plan_L', 'plan_K','plan_S','age','story', 'floor','surface','walk_min1', 'monthly_rent']]
df_o = pd.merge(df_o, ward, left_index=True, right_index=True)
df_o = df_o.dropna()
y_true = df_o['monthly_rent']
df_o.drop(['monthly_rent'], axis=1, inplace=True) 
data = df_o
```

```python
delta = clf.predict(data) - y_true
delta.name = 'delta_monthly_rent'
df = pd.concat([df, delta], axis=1)
```

```python
st = '牛込神楽坂駅'
df_out = df[df['station_name1'] == st].append(df[df['station_name2'] == st]).append(df[df['station_name3'] == st])
df_out = df_out.sort_values(by='delta_monthly_rent', ascending=False)
df_origin.ix[df_out.index]
#df_origin.head(40)
```
